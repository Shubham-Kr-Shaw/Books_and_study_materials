```python
costs = {'socks':5, 'shoes':60, 'sweater':30}
tax = 0.09
```

```python
def cal_total_bill(items_list, no_of_items_list):
    tax = 0.09
    total_cost_before_tax = 0
    for i,item in enumerate(items_list):
        total_cost_before_tax = total_cost_before_tax + costs[item] * no_of_items_list[i]
        
    total_cost_after_tax = total_cost_before_tax + total_cost_before_tax * tax
    print(total_cost_after_tax)
    
```
