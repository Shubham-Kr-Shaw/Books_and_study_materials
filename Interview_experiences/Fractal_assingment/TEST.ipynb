```python
## Count of word in a sentence.
```

```python
sent = "My name is shubham and my father's name is Suresh."
sent = sent.lower()
```

```python
words =  sent.split()
```

```python
word_counter = {}
```

```python
for word in words:
    if word not in word_counter.keys():
        word_counter[word] = 0
    
    word_counter[word] = word_counter[word]+1
```

```python
word_counter
```
